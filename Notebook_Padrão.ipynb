```python
import pandas as pd
import numpy as np
```

```python
#Gerando uma série com pandas
preço = pd.Series([3.4,4.3,5.5,12,3.5], index = ['Tijolos','Tinta','Cola','Solvente','Argamassa'])

print(preço)
```

```python
# Extraindo informações da série
print(preço ['Tijolos'])
print(preço ['Tinta'])
```

```python
# Manipulando estatísticas na tabela
print("A media de preços é:",preço.mean())

print("O desvio padrão dos preços é:",preço.std())
```

```python
#Imprimindo todos os parâmetros estatísticos da lista

print(preço.describe())
```

```python
#Elevando a série ao quadrado
quadrado_serie = preço**2
print(quadrado_serie)
```

###Trabalhando com Dataframes

```python
# Criando um Dataframes

info = pd.DataFrame({'Produto':['Tijolo','Tinta','Solvente','Telhas','Cimento'],'Preco':[3.3,4.5,12,27,45],'Qtde':[10,12,23,34,4],'Custo':[0.5,12,1.4,12,11]})
print(info)
```

```python
#Imprimindo uma coluna apenas do dataframe
info['Produto']
```

```python
#Fazendo o sort do dataframe
info.sort_values(by='Preco')
```

```python
#Ordenando o DataFrame por preço em ordem crescente
info.sort_values(by='Preco',ascending = False)
```

```python
#Filtrando o DataFrame
info['Custo']>1.2
```

```python
#Listando todos os produtos com custo maior que a condição
info[info['Custo']>1.2]
```

```python
#Refinando a lista : condição dupla
#Quais são os produtos de preço maior que 1.2 e quantidade maior que 10

info[(info['Custo']>1.2) & (info['Qtde']>10)]
```

```python
# Imputando uma nova lista de compras

df=pd.DataFrame({'Produto':['Tijolo','Tinta','Solvente','Telhas','Tijolo','Tinta'], 'Preco':[3.3,4.5,12,24,45,6],'Qtde':[10,12,23,34,4,10],'Custo':[0.5,1.2,1.4,12,12,18]},index=['Compra1','Compra2','Compra3','Compra4','Compra5','Compra6'])
df[['Produto','Preco','Qtde']]
```

```python
#comando loc - localizando linhas

df.loc['Compra1']
```

```python
#Listando apenas o produto da compra3

df.loc['Compra3']['Produto']
```

```python
#outra forma...

df[df['Produto'] ==['Tinta']]
```

```python
# Filtrando o DataFrame usando o método .query()
df.query('Produto == "Tinta"')
```

```python
#alternativa:

df.loc[(df['Produto']=='Tinta')]
```

```python
#mais uma alternativa

df.loc[(df['Produto']=='Tinta')]['Preco']
```

```python
#Busca de produtos com um preco acima de um determinado valor

df.loc[df['Preco']>7]['Produto']
```

```python
#Lista

df['Produto'].unique()
```

```python
#Listando produtos por quantidades

df['Produto'].value_counts()
```

```python
#repetindo a mesma listagem mas em % de produtos

df['Produto'].value_counts(normalize=True)
```

```python
#Calculando a média e o máximo da lista

# Média
x = df.groupby('Produto').mean()

print(x)
```

```python
# Máximo
x=df.groupby('Produto').max()

print(x)
```
