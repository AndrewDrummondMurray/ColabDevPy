```python
import matplotlib.pyplot as plt
import random as rd

# Gerando Números aleatórios com base em distribuições de probabilidades
# 1 atribuindo as duas vaŕáveis:
rand1=[]
rand2=[]

for i in range (1000):
  x= rd.uniform(-1,1) # gera uma distribuição uniforme entre (-1,1) de 1000 numeros
  rand1.append(x)

  x=rd.gauss(0,1)     # gera uma distribuição normal entre (0,1)
  rand2.append(x)

plt.figure(figsize=(10, 8)) # Create a figure to hold the subplots

# First subplot: plots of random numbers
plt.subplot(2, 1, 1) # (rows, columns, plot_number)
plt.plot(rand1,'b', label='Uniform (-1,1)')
plt.plot(rand2,'r', label='Gaussian (0,1)')
plt.title('Plots of Uniform and Gaussian Random Numbers')
plt.xlabel('Index')
plt.ylabel('Value')
plt.xlim(0, 1000)   # Set x-axis limits from 0 to 1000
plt.ylim(-4, 4)     # Set y-axis limits from -4 to 4
plt.legend()
plt.grid(True)

# Second subplot: histogram of uniform distribution
plt.subplot(2, 1, 2) # (rows, columns, plot_number)
plt.hist(rand1, bins=10, color='skyblue', edgecolor='black')
plt.title('Histogram of Uniform Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.xlim(-1.2, 1.2) # Set x-axis limits for the histogram
# plt.ylim() # Letting matplotlib automatically set y-axis limits for frequency
plt.grid(True)

plt.tight_layout() # Adjust layout to prevent overlapping elements
plt.show()
```

### Analizando as cotações de um ativo financeiro ao longo de um ano

-No caso foi fornecido um arquivo com os dados de abertura/fechamento/variação/minimo e maximo de um determinado ativo financeiro no caso a planilha UFRGS_Aaula7.xlsx
-caminho da planilha :'content/drive/MydDrive/Colab Notebooks/UFRGS/UFRGS_Aula7.xlsx
- A meta e fazer um gráfico de boxplot dstes dados


```python
# Para ler um arquivo .ods com pandas
import pandas as pd

excel_file_path = '/content/drive/MyDrive/Colab Notebooks/UFRGS/Aula07.ods'
try:
    # Ler o arquivo ODS sem cabeçalho, pois a primeira linha parece ser dados
    df_excel = pd.read_excel(excel_file_path, header=None, engine='odf')
    # Definir os nomes das colunas manualmente
    df_excel.columns = ['Date', 'Open', 'High', 'Low', 'Close']
    # Converter a coluna 'Date' para o tipo datetime
    df_excel['Date'] = pd.to_datetime(df_excel['Date'])
    print('DataFrame carregado de Excel com colunas renomeadas:')
    display(df_excel.head())
except FileNotFoundError:
    print(f"Erro: Arquivo não encontrado em {excel_file_path}")
except Exception as e:
    print(f"Ocorreu um erro ao carregar o arquivo Excel: {e}")
```

### Criando o Boxplot das Cotações

```python
import matplotlib.pyplot as plt

# Selecionando as colunas numéricas para o boxplot
columns_to_plot = ['Open', 'High', 'Low', 'Close']

plt.figure(figsize=(10, 6))
df_excel[columns_to_plot].boxplot()
plt.title('Boxplot das Cotações de Abertura, Máxima, Mínima e Fechamento')
plt.ylabel('Valor da Cotação')
plt.grid(axis='y', linestyle='--')
plt.show()
```

Este boxplot mostra a distribuição dos valores para 'Open', 'High', 'Low' e 'Close'. Cada caixa representa o intervalo interquartil (IQR), com a linha no meio sendo a mediana. Os 'whiskers' (linhas verticais) estendem-se até o menor e o maior valor dentro de 1.5 vezes o IQR a partir dos quartis, e os pontos fora dos whiskers são considerados *outliers*.

### Série Temporal do Valor Médio Mensal de Fechamento

```python
import matplotlib.pyplot as plt

# Definir 'Date' como índice para facilitar a resampleagem
df_time_series = df_excel.set_index('Date')

# Agrupar por mês e calcular a média do preço de fechamento
# Usando 'ME' (Month End) como recomendado para evitar FutureWarnings
monthly_avg_close = df_time_series['Close'].resample('ME').mean()

plt.figure(figsize=(12, 7))
monthly_avg_close.plot(kind='line', marker='o', linestyle='-', color='skyblue')
plt.title('Média Mensal do Preço de Fechamento da Ação')
plt.xlabel('Mês')
plt.ylabel('Cotação Média Mensal')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=45) # Rotacionar os rótulos do eixo x para melhor leitura
plt.tight_layout() # Ajustar o layout para evitar sobreposição
plt.show()
```

Este gráfico mostra a cotação média mensal de fechamento da ação ao longo do período disponível nos dados. Você pode observar a tendência geral e a variação mensal do preço.
